```python
#Superposition of two sine waves
import numpy as np
import matplotlib.pyplot as plt

#Parameters of Wave1
amplitude_1 = float(input('Enter the amplitude of wave 1 in m: '))
frequency_1 = float(input('Enter the frequency of wave 1 in Hz: '))
phase_1 = float(input('Enter the phase of wave 1 in radians: '))
duration_1= float(input('Enter the duration of wave 1 you want to view in seconds: '))
sampling_1= int(input('Enter the sampling rate in Hz: '))

#Parameters of Wave2
amplitude_2 = float(input('Enter the amplitude of wave 2 in m: '))
frequency_2 = float(input('Enter the frequency of wave 2 in Hz: '))
phase_2 = float(input('Enter the phase of wave 2 in radians: '))
duration_2 = float(input('Enter the duration of wave 2 you want to view in seconds: '))
sampling_2= int(input('Enter the sampling rate in Hz: '))

#time vector for wave one
t_1 = np.linspace(0,duration_1,int(sampling_1*duration_1), endpoint=False)
#time vector for wave two
t_2 = np.linspace(0,duration_2,int(sampling_2*duration_2), endpoint= False)

#Wave 1 Sine Wave
y_1 = amplitude_1 * np.sin(2 * np.pi * frequency_1 * t_1 + phase_1)

#Wave 2 Sine Wave
y_2 = amplitude_2 * np.sin(2 * np.pi * frequency_2 * t_2 + phase_2)

#Wave 3 Superposed Wave
y= y_1+ y_2

#Plotting the waves
#The given set of programs will work properly for t_1 = t_2
plt.figure(figsize=(16,9))
plt.plot(t_1,y_1, label='Wave 1', color= 'red')
plt.plot(t_2,y_2, label='Wave 2', color = 'blue')
plt.plot(t_1,y, label = 'Superposed Wave', color = 'green' )
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Superposition of two waves')
plt.legend()
plt.grid(True)
plt.show()
```
