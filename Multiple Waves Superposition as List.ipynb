```python
#Superposition in lists
#Program that accepts the input into lists and iterates it two times in a for loop to do superposition
import numpy as np
import matplotlib.pyplot as plt

#sine function generator
def sine_wave(amplitude,frequency,duration,phase,sampling):
  t= np.linspace(0,duration,int(sampling*duration), endpoint = False)
  y= amplitude*np.sin(2*np.pi*frequency*t + phase)
  return t,y

parameters=[]
for i in  range(1,3):
  print(f'Input Parameters for Wave{i}')
  amplitude =float(input('Enter Amplitude in m: '))
  frequency =float(input('Enter Frequency in Hz: '))
  duration =float(input('Enter Duration you want to see the wave for in s: '))
  phase =float(input('Enter Phase in rad: '))
  sampling =float(input('Enter Sampling Rate in Hz:'))
  parameters.append((amplitude,frequency,duration,phase,sampling))

t_1,y_1=sine_wave(parameters[0][0],parameters[0][1],parameters[0][2],parameters[0][3],parameters[0][4])
t_2,y_2=sine_wave(parameters[1][0],parameters[1][1],parameters[1][2],parameters[1][3],parameters[1][4])

y= y_1+y_2

plt.figure(figsize=(16,9))
plt.plot(t_1,y_1,label='Wave 1')
plt.plot(t_2,y_2,label='Wave 2')
plt.plot(t_1,y,label='Superposition')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Superposition of Sine Waves')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Superposition of multiple waves
#This program intends to supply the user with the option of superposing more than two waves
import numpy as np
import matplotlib.pyplot as plt

#sine function generator
def sine_wave(amplitude,frequency,duration,phase,sampling):
  t= np.linspace(0,duration,int(sampling*duration), endpoint = False)
  y= amplitude*np.sin(2*np.pi*frequency*t + phase)
  return t,y

waves_added= int(input('Enter the number of waves you want to add: '))
parameters=[]
for i in  range(1,waves_added+1):       #We add 1 to ensure that the for loop doesn't end iterating one step before the number of waves  
  print(f'Input Parameters for Wave{i}')
  amplitude =float(input('Enter Amplitude in m: '))
  frequency =float(input('Enter Frequency in Hz: '))
  duration =float(input('Enter Duration you want to see the wave for in s: '))
  phase =float(input('Enter Phase in rad: '))
  sampling =float(input('Enter Sampling Rate in Hz:'))
  parameters.append((amplitude,frequency,duration,phase,sampling))

plt.figure(figsize=(16,9))
y_sup= 0 #SuperpositionValue
#for loop that sets superposition value initially at i and then adds to it after every iteration
for i in range(len(parameters)):
  t,y=sine_wave(parameters[i][0],parameters[i][1],parameters[i][2],parameters[i][3],parameters[i][4])
  plt.plot(t,y,label=f'Wave {i+1}')
  y_sup+= y

plt.plot(t,y_sup,label='Superposition')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Superposition of Sine Waves')
plt.grid(True)
plt.legend()
plt.show()
```
