```python
#Sine Wave Generation
import numpy as np
import matplotlib.pyplot as plt

# Parameters
frequency = 1  # Frequency in Hz
amplitude = 1  # Amplitude of the sine wave
duration = 2   # Duration in seconds
sampling_rate = 50000 # Sampling rate in Hz

# Time vector from 0 to duration with the given sampling rate
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate sine wave
sine_wave = amplitude * np.sin(2 * np.pi * frequency * t)

# Plot the sine wave
plt.figure(figsize=(16,9))
plt.plot(t, sine_wave)
plt.title(f'Sine Wave - {frequency} Hz')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
#Sine Wave Generation
import numpy as np
import matplotlib.pyplot as plt

# Parameters
frequency = 1  # Frequency in Hz
amplitude = 1  # Amplitude of the sine wave
duration = 2   # Duration in seconds
sampling_rate = 5 # Sampling rate in Hz

# Time vector from 0 to duration with the given sampling rate
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate sine wave
sine_wave = amplitude * np.sin(2 * np.pi * frequency * t)

# Plot the sine wave
plt.figure(figsize=(16,9))
plt.plot(t, sine_wave)
plt.title(f'Sine Wave - {frequency} Hz')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

On reducing the sampling rate, the number of points generated becomes less, which does not give a wave as an output

### Time vector from 0 to duration with the given sampling rate
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

### Generate sine wave
sine_wave = amplitude * np.sin(2 * np.pi * frequency * t)

We know $y = A\sin(2\pi ft)$ <br>
A is the amplitude provided(1)<br>Frequency is provided.<br>t is calculated

### Calculation of t
A wave requires multiple datapoints. In the given expression, A and f are constant. Multiple values of t for the requisite datapoints are generated.<br>
The datapoints need to be evenly spaced, which is achieved by the linspace function generating an array<br>0 is the initial point<br>duration, set as 2, is the final point<br>sampling point(50000)*duration(2) gives 100000 data points between 0 and 2<br> endpoint set as false ensures that the 2 is not included

```python
#Program to take User Input
import numpy as np
import matplotlib.pyplot as plt

# User inputs
frequency = float(input("Input frequency in Hz: "))  # Frequency in Hz
amplitude = float(input("Input amplitude: "))  # Amplitude of the sine wave
duration = float(input("Input duration in seconds to be visible on the plot: "))  # Duration in seconds
sampling_rate = int(input("Enter the sampling rate in Hz: "))  # Sampling rate in Hz

# Time vector from 0 to duration with the given sampling rate
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate sine wave
sine_wave = amplitude * np.sin(2 * np.pi * frequency * t)

# Plot the sine wave
plt.figure(figsize=(10, 4))
plt.plot(t, sine_wave)
plt.title(f'Sine Wave - {frequency} Hz')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
#Sine Wave with phase difference and User Input
import numpy as np
import matplotlib.pyplot as plt

# User inputs
frequency = float(input("Input frequency in Hz: "))  # Frequency in Hz
amplitude = float(input("Input amplitude: "))  # Amplitude of the sine wave
duration = float(input("Input duration in seconds to be visible on the plot: "))  # Duration in seconds
sampling_rate = int(input("Enter the sampling rate in Hz: "))  # Sampling rate in Hz
phase_difference = float(input("Input phase difference in radians: "))  # Phase difference in radians

# Time vector from 0 to duration with the given sampling rate
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate sine wave
sine_wave = amplitude * np.sin(2 * np.pi * frequency * t + phase_difference)

# Plot the sine wave
plt.figure(figsize=(10, 4))
plt.plot(t, sine_wave)
plt.title(f'Sine Wave - {frequency} Hz')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
